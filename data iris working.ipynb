```python
import numpy as np
url=(r"C:/users/LENOVO/Desktop/pdf&word/python/libraries/irisdataset.csv")
iris1=np.genfromtxt(url,delimiter=',',dtype='float',usecols=[0,1,2,3])
names=('sepallength','sepalwidth','petallength','petalwidth','species')
np.any(np.isnan(iris1))
iris1[np.isnan(iris1)]=0 #because isnan is character and will not do mathematical operations
print("mean of sepallength is: ",np.mean(iris1[1:,[0]]))
print("mean of petallength is: ",np.mean(iris1[1:,[2]]))
print("median of sepallength is: ",np.median(iris1[1:,[0]]))
print("median of petallength is: ",np.median(iris1[1:,[2]]))
print("standard deviation of sepallength is: ",np.std(iris1[1:,[0]]))
print("standard deviation of petallength is: ",np.std(iris1[1:,[2]]))

print("petallength greater than 1.5 is: ",iris1[iris1[0:,2]>1.5])
```
